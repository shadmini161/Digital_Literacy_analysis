# gender distribution

```python
plt.figure(figsize=(4,4))
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show
```

```python
# from above chart we have analysed that the number of females in the data is more than the number of males.
```

# impact of Employment_Status on Basic_Computer_Knowledge_Score and Mobile_Literacy_Score  

```python
gb=df.groupby("Education_Level").agg({"Mobile_Literacy_Score":'mean',"Basic_Computer_Knowledge_Score":'mean'})
print(gb)
```

```python
plt.figure(figsize=(7,4))
sns.heatmap(gb)
plt.title("Reletionship b/w edu lvl and mobile&comp literacy scores")
plt.show()
```

```python
#from the above chart we conclude that the education level has a impact on the mobile and computer knowledge scores.
```

# impact on Mobile_Literacy_Score based on Modules_Completed and Average_Time_Per_Module and Quiz_Performance.  

```python
gb1=df.groupby("Mobile_Literacy_Score").agg({"Modules_Completed":'mean',"Average_Time_Per_Module":'mean',"Quiz_Performance":'mean'})
print(gb1.head())
```

```python
gb1 = df.groupby("Mobile_Literacy_Score").agg({
    "Modules_Completed": 'mean',
    "Average_Time_Per_Module": 'mean',
    "Quiz_Performance": 'mean'
})
plt.figure(figsize=(10, 7))
plt.plot(gb1.index, gb1["Modules_Completed"], label="Modules Completed", marker='o')
plt.plot(gb1.index, gb1["Average_Time_Per_Module"], label="Average Time per Module", marker='s')
plt.plot(gb1.index, gb1["Quiz_Performance"], label="Quiz Performance", marker='^')
plt.xlabel("Mobile Literacy Score")
plt.ylabel("Mean Values")
plt.title("Line Plot of Aggregated Data by Mobile Literacy Score")
plt.legend()
plt.grid(True)
plt.show()
```
